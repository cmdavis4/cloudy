```python
import matplotlib.pyplot as plt
import pickle as pkl
from pathlib import Path

import figure_data as fd
```

```python
# Start the figure; this resets the data from any previous figures,
# as well as installing the necessary hooks into matplotlib if they haven't been already
fd.start_figure("F01_two_lines")
# This does return a FigureData object, but with the usage pattern here you don't need
# to do anything with it; like a figure in matplotlib, it keeps track of which one
# is active internally

xs = list(range(10))
fig, ax = plt.subplots()

# Can plot things using either the plt. functions;
plt.plot(xs, list(range(10)), name="positive_line")
# Or using methods of axes objects
ax.plot(xs, list(range(0, -10, -1)), name="negative_line")
# In both cases, the only change from normal is the "name" kwarg; the actual data
# saved to disk is a dictionary where the keys are these "name" arguments and the
# values are the data passed to the function/method

# The data that gets saved is whatever unnamed arguments you pass to the
# function/method; in some cases, this might save something like the line marker
# symbol, but with standard usage it should save *at least* the actual data
# passed to the functions

# Save the data
fd.save_figuredata("F01_data.pkl")

# Save the figure as normal
fig.savefig("F01_two_lines.png")
```
